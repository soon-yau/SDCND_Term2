```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
dataset=pd.read_table('output1.txt')
dataset_radar=dataset[dataset.sensor_type=='R']
dataset_laser=dataset[dataset.sensor_type=='L']

f=plt.figure()
dataset_radar['NIS'].plot(ax=f.gca())
df=pd.DataFrame(np.full((1200),7.815))
df.plot(ax=f.gca(), legend=None)

plt.title('Radar NIS')
plt.show()

f2=plt.figure()
dataset_laser['NIS'].plot(ax=f2.gca())
df2=pd.DataFrame(np.full((1200),5.991))
df2.plot(ax=f2.gca(), legend=None)
plt.title('Laser NIS')
plt.show()
```
